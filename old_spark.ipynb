```python
import nfl_data_py as nfl
```

```python
import pandas 
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('nfl').getOrCreate()
```

```python
df = nfl.import_depth_charts([2022])
```

```python
df[(df['position']=='QB') & (df['club_code'] == 'ATL')].drop_duplicates(subset = 'gsis_id')
```

```python
teams = nfl.import_team_desc()
```

```python
teams
```

```python
teams = spark.createDataFrame(teams.drop(['team_color3', 'team_color4'], axis=1))
```

```python
players = spark.createDataFrame(df)
```

```python
df
```

```python
teams_and_players = players.join(teams, players.club_code == teams.team_abbr, how='inner')
```

```python
weekly = spark.createDataFrame(nfl.import_weekly_data(years= [2022]))
```

```python
len(weekly.columns) + len(players.columns)+ len(teams.columns)
```

```python
players_and_weekly_stats = teams_and_players.join(weekly, weekly.player_id == teams_and_players.gsis_id, how='inner')
```

```python
players_and_weekly_stats.toPandas()
```

```python
teams_and_players.createOrReplaceTempView('players_teams')
```

```python
spark.sql("select * from players_teams where (club_code = 'ARI' or club_code = 'NE') and formation = 'Offense' and week = '14'").toPandas()
```

```python
spark.sql('select distinct club_code from players_teams').toPandas()
```
